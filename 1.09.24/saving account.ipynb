```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#setup the simulation with constant D, W values

r = 0.03
D = 100
W = 50
P = 100
```

```python
n_iters = 36
A = [P]
dt = 1

for i in range(1, n_iters):
    Ai = A[i - 1] + dt * (r * A[i - 1] + D - W)
    A.append(Ai)
```

```python
_ = plt.figure(figsize=(10, 5))
plt.plot(range(n_iters), A, c = "green", linestyle = "-.")
plt.scatter(range(n_iters), A, s = 50)
plt.xlabel("Month #")
plt.ylabel("Accout value EUR")
plt.title("Account value over time")
```

```python
_ = plt.figure(figsize=(10, 5))
plt.plot(range(n_iters), A, c = "green", linestyle = "-.")
plt.scatter(range(n_iters), A, s = 50)
plt.xlabel("Month #")
plt.ylabel("Accout value EUR")
plt.title("Account value over time with random withdrawals and deposits")
```

```python
#SETUP WITH MAX SPENDINGS

n_iters = 120
dt = 1
P = 100
D = np.random.random(n_iters)
W = np.random.random(n_iters)
```

```python
A = [P]

for i in range(1, n_iters):
    Ai = A[i - 1] + dt * (r * A[i - 1] + D[i - 1] * A[i - 1] - W[i - 1] * A[i - 1])
    A.append(Ai)
```
