===
Taylor series
---

```python
import numpy as np
import matplotlib.pyplot as plt

from math import factorial
from math import pow
```

```python
def cosapprox(x, power):
    cos = 1

    powers = list(range(2, power, 2))

    for j, i in enumerate(powers):
        term = np.power(x, i) / factorial(i)
        term = np.power(-1, j + 1) * term

        cos = cos + term
    
    return cos
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 200)
cos1 = np.cos(x)
cos2 = cosapprox(x, 4)
cos3 = cosapprox(x, 20)
cos4 = cosapprox(x, 100)
```

```python
_ = plt.figure(figsize=(10, 5))
plt.plot(x, cos1, label='np', linestyle = "--")
plt.plot(x, cos2, label='taylor 4', linestyle = "-")
plt.plot(x, cos3, label='taylor 20', linestyle = "-")
plt.plot(x, cos4, label='taylor 1000', linestyle = "-")
plt.legend()
plt.show()
```
