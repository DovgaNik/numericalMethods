```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    y = -5 * x + 20
    return y
```

```python
x = np.linspace(-5, 5, 20)
y = f(x)
_ = plt.figure()

plt.plot(x, y)
plt.scatter(x, y, s = 5)
plt.hlines(0, -5, 5, linestyles='dashed')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
a = 2
b = 5
N = 100

def bisection (f, a, b, N):
    if f(a) * f(b) >= 0:
        print('Method fails, please select a different interval')

    a_n = a
    b_n = b

    for n in range(1, N + 1):
        m = (a_n + b_n) / 2
        f_m = f(m)
        if f(a_n) * f_m < 0:
            a_n = a_n
            b_n = m
        elif f(m) * f(b_n) < 0:
            a_n = m
            b_n = b_n
        elif f_m == 0:
            print(f'Found exact solution x0 = {m}')
            return m
        else:
            print('Method fails')
    return (a_n + b_n) / 2

bisection(f, a, b, N)
```

```python
#Newnon's method

def Dh(x):
    y = 2*x - 1
    return y
```

```python
x = np.linspace(-3, 3)
y = h(x)
Dy = Dh(x)

_ = plt.figure(figsize = (10, 8))
plt.plot(x, y, label = 'h(x)')
plt.plot(x, Dy, label = "h'(x)")
plt.xlabel('x')
plt.legend()
plt.show()
```

```python
def newton(f, Df, x0, eps, max_iter):
    xn = x0

    for n in range(max_iter):
        fxn = f (xn)
        if abs (fxn) < eps:
            return xn
        Dfxn = Df(xn)

        if Dfxn == 0:
            print("Zero derivative. No solution available")
            return None
        
        xn = xn - fxn / Dfxn

    print("Exceeded maximum iterations. No solution found")
```

```python
x1 = newton(h, Dh, x0 = -1, eps = 0.0001, max_iter=1000)
x2 = newton(h, Dh, x0 = 2, eps = 0.0001, max_iter=1000)

print(x1, x2)
```
