C(x) = ax^3 - bx^2 + cx + d
C'(x) = a * 3 * x^2 - b * 2 *x + c

a, b, c, d are known

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0.01
b = 1.2
c = 5.0
d = 100.0

start_x = 10.0
learn_rate = 0.01
N = 1000
```

```python
def cost(x):
    return a * x**3 - b * np.square(x) + c * x + d
```

```python
def cost_gradient(x):
    return 3 * a * np.square(x) - 2 * b * x + c
```

```python
# Just to check that derivative is correct
x = np.linspace(-100, 100)
_ = plt.plot(cost_gradient(x), color =  "red")
plt.plot(cost(x))
plt.show()
```

```python
def gradient_descent(start_x, learning_rate, iterations):
    history = []

    x = start_x
    for i in range(iterations):
        x_tmp = x
        x = x - learning_rate * cost_gradient(x)
        history.append([x_tmp, x])
    return history
    
```

```python
x = np.linspace(-10, 10)


res_x = gradient_descent(x, learn_rate, N)[-1][0]
res_y = gradient_descent(x, learn_rate, N)[-1][1]
_ = plt.plot(cost_gradient(x), color =  "red", label = "Cost gradient")
plt.plot(cost(x), label = "Function itself")
plt.scatter(gradient_descent(x, learn_rate, N)[0], gradient_descent(x, learn_rate, N)[1], color = "orange", label = "Gradient descent")
plt.scatter(res_x, res_y)
plt.legend()
plt.grid()
plt.show()
```
